```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  
```

```python
df = pd.read_csv("dataset.csv")
df
```

```python
import pandas as pd

missing_counts = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_summary = pd.DataFrame({
    "missing_count": missing_counts,
    "missing_percent": missing_percent
})

missing_summary.sort_values("missing_percent", ascending=False)
```

```python
df = df.dropna(subset=["rainfall"])
```

```python
weather_cols = ["avg_temp", "min_temp", "max_temp", "wind_speed", "air_pressure"]

df = df.sort_values(["station_name", "date_of_record"])

df[weather_cols] = df.groupby("station_name")[weather_cols].transform(
    lambda x: x.interpolate(method="linear").ffill().bfill()
)
df
```

```python
import pandas as pd

missing_counts = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_summary = pd.DataFrame({
    "missing_count": missing_counts,
    "missing_percent": missing_percent
})

missing_summary.sort_values("missing_percent", ascending=False)
```

```python
df["temp_k"] = df["avg_temp"] + 273.15
df["pressure_pa"] = df["air_pressure"] * 100

R = 287
df["air_density"] = df["pressure_pa"] / (R * df["temp_k"])
df[["temp_k", "pressure_pa", "air_density"]]
```

```python
df["lat_rad"] = np.radians(df["latitude"])

df["u_velocity"] = df["wind_speed"] * np.cos(df["lat_rad"])
df["v_velocity"] = df["wind_speed"] * np.sin(df["lat_rad"])
df[["u_velocity", "v_velocity", "lat_rad"]]
```

```python
df["pressure_gradient"] = df.groupby("station_name")["air_pressure"].diff().fillna(0)
df["pressure_gradient"]
```

```python
df
```

```python
df["du_dx"] = df.groupby("station_name")["u_velocity"].diff().fillna(0)
df["dv_dy"] = df.groupby("station_name")["v_velocity"].diff().fillna(0)

df["divergence"] = df["du_dx"] + df["dv_dy"]
df[["du_dx", "dv_dy", "divergence"]]
```

```python
df["dv_dx"] = df.groupby("station_name")["v_velocity"].diff().fillna(0)
df["du_dy"] = df.groupby("station_name")["u_velocity"].diff().fillna(0)

df["vorticity"] = df["dv_dx"] - df["du_dy"]
df['vorticity']
```

```python
omega = 7.2921e-5

df["coriolis"] = 2 * omega * np.sin(df["lat_rad"])
df[["lat_rad", "coriolis"]]
```

```python
df["kinetic_energy"] = 0.5 * df["air_density"] * (df["wind_speed"]**2)
df[["air_density", "wind_speed", "kinetic_energy"]]
```

```python
df["temp_gradient"] = df.groupby("station_name")["avg_temp"].diff().fillna(0)
df[["avg_temp", "temp_gradient"]]
```

```python
df["rain_lag_1"] = df.groupby("station_name")["rainfall"].shift(1).fillna(0)
df["rain_lag_3"] = df.groupby("station_name")["rainfall"].shift(3).fillna(0)
df["rain_lag_7"] = df.groupby("station_name")["rainfall"].shift(7).fillna(0)
df["rain_lag_30"] = df.groupby("station_name")["rainfall"].shift(30).fillna(0)
df[["rainfall", "rain_lag_1", "rain_lag_3", "rain_lag_7", "rain_lag_30"]]
```

```python
df["date_of_record"] = pd.to_datetime(df["date_of_record"], errors="coerce")
df["month"] = df["date_of_record"].dt.month

df["month_sin"] = np.sin(2*np.pi*df["month"]/12)
df["month_cos"] = np.cos(2*np.pi*df["month"]/12)
df[["month", "month_sin", "month_cos"]] 
```

```python
import pandas as pd

missing_counts = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_summary = pd.DataFrame({
    "missing_count": missing_counts,
    "missing_percent": missing_percent
})

missing_summary.sort_values("missing_percent", ascending=False)
```

```python
df.to_csv("processed_weather_data.csv", index=False)
```
